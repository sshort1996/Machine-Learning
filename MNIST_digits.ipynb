## Recognising MNIST digits

First task here is to get our hands on the MNIST data and see what they look like

julia library MLDatasets allows this easily it seems

```julia
using Pkg
Pkg.add("MLDatasets")
```

```julia
using MLDatasets

training_x, training_y = MNIST.traindata()
test_x, test_y = MNIST.testdata()
```

```julia
@show length(training_x)
@show length(training_y)
@show length(test_x)
@show length(test_y)
```

```julia
using Plots
```

```julia
p1 = heatmap(transpose(test_x[:,:,9]), c =:grays)
p2 = heatmap(transpose(test_x[:,:,10]), c =:grays)
p3 = heatmap(transpose(test_x[:,:,11]), c =:grays)
p4 = heatmap(transpose(test_x[:,:,12]), c =:grays)

plot(p1, p2, p3, p4, layout = @layout([A B C D]), size = (900,200), c =:grays, yflip = true)
```

neato. 

### Set up cost function 

```julia
function cost(error::Array{Float64,1})
    C = 1/(2*length(error)) * sum(abs2.(error))
    return C
end
```

```julia
# random test
error = rand(10)
@show cost(error)
```
